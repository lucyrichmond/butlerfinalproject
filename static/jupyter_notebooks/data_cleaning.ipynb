```python
import pandas as pd
import json
```

```python
# Path to the csv
movie_metadata = "data/movies_metadata.csv"
opening = "data/opening.csv"

# make dataframe from csv
metadata_df = pd.read_csv(movie_metadata, low_memory = False)
metadata_df.head()
opening_df = pd.read_csv(opening)
#opening_df.head()
```

```python
opening_df = opening_df.drop(columns=['Unnamed: 0'])
opening_df
```

```python
# convert release_date to datetime
metadata_df['release_date'] = pd.to_datetime(metadata_df['release_date'], errors='coerce')
metadata_df.head()

```

```python
# add column with year only
year = metadata_df['release_date'].dt.year
metadata_df['Release Year'] = year
metadata_df = metadata_df.fillna(0)
metadata_df['Release Year'] = metadata_df['Release Year'].astype(int)

metadata_df.head()
metadata_df.shape
```

```python
# add column with movie title and release year
metadata_df["Movie & Year"] = metadata_df["title"].astype(str) + " " + "(" + metadata_df["Release Year"].astype(str) + ")"
metadata_df.head()
```

```python
#dropping unecessarry columns
metadata_df = metadata_df.drop(columns = ['adult', 'belongs_to_collection', 'homepage', 'overview', 'tagline', 'poster_path'])
metadata_df
```

```python
metadata_df = pd.merge(opening_df, metadata_df, on="Movie & Year", how="right")
metadata_df.head()
```

```python
metadata_df.shape
```

```python
metadata_df.to_csv('data/metadata.csv')
```

```python
genres = metadata_df['genres']
prod_companies = metadata_df['production_companies']
```

```python
ids = genres.apply(lambda x: [y['id'] for y in json.loads(x.replace("'", '"')) if len(y) > 0])
names = genres.apply(lambda x: [y['name'] for y in json.loads(x.replace("'", '"')) if len(y) > 0])
names
```

```python
def getFirst(x):
    if len(x) > 0:
        return x[0]
    else:
        return []
    
firstIds = ids.apply(getFirst)
firstNames = names.apply(getFirst)
firstNames
```

```python
metadata_df['Genre Ids'] = firstIds
metadata_df['Genre Names'] = firstNames
metadata_df
```

```python
metadata_df = metadata_df.loc[(metadata_df["original_language"] == "en")]
metadata_df
```

```python
def idWithTry(x):
    try:
        return [y['id'] for y in json.loads(x.replace("'", '"')) if len(y) > 0]
    except:
        return None
    
def nameWithTry(x):
    try:
        return [y['name'] for y in json.loads(x.replace("'", '"')) if len(y) > 0]
    except:
        return None
        

production_companies = metadata_df["production_companies"]
production_ids = production_companies.apply(idWithTry)
production_names = production_companies.apply(nameWithTry)

# final_prod_ids = production_ids.apply(getFirst)
# final_prod_names = production_names(getFirst)
# final_prod_ids

production_ids = production_ids.dropna()
production_names = production_names.dropna()
```

```python
company_ids = production_ids.apply(getFirst)
company_names = production_names.apply(getFirst)
```

```python
new_df = pd.DataFrame({"production_ids": production_ids, "genre ids": firstIds, "production names": production_names})
new_df
```

```python
metadata_df['Company Ids'] = company_ids
metadata_df['Company Names'] = company_names
metadata_df
```

```python
metadata_df = metadata_df.drop(columns=['spoken_languages', 'Release Year_y', 'status', 'video'])
metadata_df = metadata_df.rename(columns={'Release Year_x': 'Release Year'})
metadata_df.head()
```
